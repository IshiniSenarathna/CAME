```python
import csv
import os

class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id.strip()
        self.name = name.strip()


students = []
while True:
    student_id = input("Enter Student ID (or type 'exit' to finish): ").strip()
    if student_id.lower() == 'exit':
        break
    if not student_id:
        print("Student ID cannot be blank. Try again.")
        continue
    name = input("Enter Student Name: ").strip()
    if not name:
        print("Student Name cannot be blank. Try again.")
        continue
    student = Student(student_id, name)
    students.append(student)
    print(f"Added: {student.student_id} - {student.name}")


csv_file = 'students.csv'
write_header = not os.path.exists(csv_file)

to_write = [[student.student_id, student.name] for student in students]
with open(csv_file, 'a', newline='') as file:
    writer = csv.writer(file)
    if write_header:
        writer.writerow(['student_id', 'name'])
    writer.writerows(to_write)

print(f"Saved {len(students)} record(s) to {csv_file}. Goodbye!")
```

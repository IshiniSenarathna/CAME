```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]
    left_sorted = merge_sort(left_half)
    right_sorted = merge_sort(right_half)
    return merge(left_sorted, right_sorted)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

def print_array(arr):
    print("[", end="")
    for i in range(len(arr)):
        print(arr[i], end="")
        if i < len(arr) - 1:
            print(", ", end="")
    print("]")

test_cases = [
    [3, 0, -2, 7, 4, -5, 0],
    [5, 1, 3, 3, 2, 5, 1],
    [-4, -1, 0, 2, 6, 9],
    [10, 7, 5, 3, 1, 0, -1],
    [0, 0, -1, -1, -3, 0, 2],
    [12, -8, 4, 0, 19, 7, -3, 7, 2, -10, 5, 5, 14, 0, -1, 9, 3, -6, 8, 11]
]

print("Merge Sort Challenge Results\n")
for idx, arr in enumerate(test_cases, start=1):
    print(f"Test {idx}:")
    print("Input: ", end="")
    print_array(arr)
    sorted_arr = merge_sort(arr)
    print("Sorted:", end=" ")
    print_array(sorted_arr)
    print()
```

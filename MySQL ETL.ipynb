```python
import os
import requests
import mysql.connector
from mysql.connector import Error


DB_CONFIG = {
    'host': os.getenv('DB_HOST'),
    'port': int(os.getenv('DB_PORT', 3306)),
    'user': os.getenv('DB_USER'),
    'password': os.getenv('DB_PASSWORD'),
    'database': os.getenv('DB_NAME')
}


url = 'https://microsoftedge.github.io/Demos/json-dummy-data/64KB.json'
response = requests.get(url)
data = response.json()  


def etl_load_to_mysql(data):
    try:
        conn = mysql.connector.connect(**DB_CONFIG)
        cursor = conn.cursor()
        cursor.execute('TRUNCATE TABLE persons')  # Remove previous data
        
        insert_sql = 'INSERT INTO persons (id, name, language, bio, version) VALUES (%s, %s, %s, %s, %s)'
        batch = [
            (p['id'], p['name'], p['language'], p['bio'], str(p['version']))
            for p in data
        ]
        cursor.executemany(insert_sql, batch)
        conn.commit()
        print(f"Inserted {cursor.rowcount} records into persons")
    except Error as e:
        print(f"Database error: {e}")
    finally:
        cursor.close()
        conn.close()

etl_load_to_mysql(data)
```
